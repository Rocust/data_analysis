```python
df = pd.read_csv('member.txt')
```

```python
group_gender = df.groupby('gender')
```

```python
group_gender.groups
```

```python
group_gender.size()
```

```python
gender_cnt = pd.DataFrame(data=group_gender.size(), columns=['count']).reset_index()
```

```python
gender_cnt
```

```python
for name, group in group_gender:
    print(name, ":", str(len(group)))
    print(group)
    print()
```

##### 1. class를 grouping
##### 2. 출력 포맷
#####    class 이름:숫자
##### 3. 출력 데이터를 새로운 DataFrame으로 생성

```python
group_class = df.groupby('class')
group_class.groups
```

```python
group_class.size()
```

```python
for name, group in group_class:
    print(name, ":", str(len(group)))
```

```python
class_cnt = pd.DataFrame({'count':group_class.size()}).reset_index()
class_cnt
```

```python
group_age = df.groupby('age')
group_age.groups
```

```python
for name, group in group_age:
    print(name, ":", str(len(group)))
```

```python
group = df.groupby(['class', 'gender'])
group.size()
pd.DataFrame(data=group.size(), columns=['count']).reset_index()
```
