```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
product_list = [{'name':'Mouse', 'price':100, 'company':'A'},
               {'name':'SSD', 'price':200, 'company':'B'},
               {'name':'CPU', 'price':300, 'company':'C'}]
df = DataFrame(product_list, columns=['name', 'price', 'company'])
```

```python
df
```

```python
df['expires'] = 5
df
```

```python
series = Series([10, 2, 2])
df['count'] = series
df
```

```python
df['total'] = df['price'] * df['count']
df
```

```python
discount = []
for row in df['total']:
    if row >= 1000:
        discount.append('10%')
    elif row >= 500:
        discount.append('5%')
    else:
        discount.append('0%')
df['discount']  =discount
df
```

```python
def is_wireless(row):
    if row == 'Mouse':
        return True
    else:
        return False

df['wireless'] = df.name.apply(is_wireless)
df
```
