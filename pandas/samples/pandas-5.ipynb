```python
list = [
    {'join_date':'2004-08-09'},
     {'join_date':'2009-08-09'},
     {'join_date':'2012-08-09'},
     {'join_date':'2013-08-09'},
     {'join_date':'2017-08-09'}
]
```

```python
df = pd.DataFrame(list)
```

```python
df
```

```python
def get_continuous_year(year, current_year):
    _year = year.split('-')[0]
    return current_year - int(_year)
```

```python
df['continuous_year'] = df['join_date'].apply(get_continuous_year, current_year=2021)
```

```python
df = pd.read_csv('customer.txt')
# Director -> 이사, Manager -> 과장, Staff -> 사원
```

```python
def changeRole(role):
    if role == 'Director':
        return '이사'
    elif role == 'Manager':
        return '과장'
    elif role == 'Staff':
        return '사원'
    else:
        return '-'
```

```python
df['converted_role'] = df['role'].apply(changeRole)
```

```python
df['converted_role2'] = df['role'].map({'Director':'이사','Manager':'과장','Staff':'사원'})
```
